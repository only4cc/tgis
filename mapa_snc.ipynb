# Mapa Saldo No castigado

```python
import pandas as pd
import json
import folium
from folium import plugins

from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

### Datos de Lista Unica

```python
filename = 'recoline_socofin_lu_SOCO_F2_F3_2019_01020304050607.zip'

final_path = '/current/eContact/Recoline/data/F2_F3_2019/'
####  Importante omitir la unidad de disco: "C:""
# notebook vaio
local_path = '/Users/Julio/Dropbox' 
# notebook econtact
#local_path = '/Users/jtrumper'
%timeit
df = pd.read_csv(local_path+final_path+filename)
```

```python
df.info()
```

```python
df = df[['rut_cliente','saldo_no_castigado', 'fecha_generacion']]
```

```python
def getmonth(f):
    return f[3:5]
df['mes'] = df['fecha_generacion'].apply(getmonth)
ultimo_mes = df['mes'].max()

# Selecciona solo filas del ultimo_mes
df = df[ (df['mes'] == ultimo_mes) ]
```

### Datos Socio-Economicos

```python
filename_se = local_path + '/current/eContact/Recoline/data/DATOS_CLI/datos_clientes_2019.zip'
df_se = pd.read_csv(filename_se)
```

```python
df_se.info()
```

```python
dfm = pd.merge(df, df_se, how='left', # how='inner', 
              #on='rut_cliente', 
              left_on='rut_cliente', right_on='rut',
              left_index=True, right_index=False, sort=True,
              suffixes=('', '_se'), copy=True, indicator=True,
              validate=None)
```

```python
dfm.info()
```

```python
#select max(mes),rut,dv,saldo_no_castigado,comuna from dfm; 
#select saldo_no_castigado as sum_snc, rut as num_ruts, comuna from dfm; 

q = '''
select sum(saldo_no_castigado) as sum_snc, count(rut)+1 as num_ruts, comuna from dfm group by comuna; 
'''
```

```python
ndf = pysqldf(q)
```

```python
ndf.head(3)
```

```python
fn = 'data/' + 'chile.json'
with open(fn, 'r') as datafile:
    data = json.load(datafile)

dfch = pd.DataFrame(data) 
```

```python
dfch.head(2)
```

```python
print(ndf['comuna'].count(), dfch['name'].count())
```

```python
dfm2 = pd.merge(ndf, dfch, how='left', 
              left_on='comuna', right_on='name',
              left_index=True, right_index=False, sort=True,
              suffixes=('', '_se'), copy=True, indicator=True,
              validate=None)
```

```python
dfm2['snc_norm'] = dfm2.apply(lambda x: (x.sum_snc / x.num_ruts), axis=1)
```

```python
dfm2.head(3)
```

```python
dfm2.dropna(inplace=True);
```

```python
p = [dfm2['lat'][:1], dfm2['lng'][:1]] 
```

```python
m = folium.Map(
    location=p,
    zoom_start=8
)

tooltip = 'click para info!'
for i in range(1,len(dfm2)):
    x = float(dfm2.iloc[i]['lat'])
    y = float(dfm2.iloc[i]['lng'])
    p = [x, y]     
    #s_sum_snc = dfm2.iloc[i]['sum_snc'].to_string()
    folium.Marker(p, popup='<i>'+dfm2.iloc[i]['comuna']+'</i>', tooltip=tooltip).add_to(m)
    #folium.CircleMarker(p, radius=3, weight=2, color='red', fill_color='red', fill_opacity=.5).add_to(m)
```

```python
m
```

```python
mh = folium.Map(
    location=p,
    zoom_start=8
)

for i,row in dfm2.iterrows():
    folium.CircleMarker((row.lat,row.lng), radius=1, weight=2, color='red', fill_color='red', fill_opacity=.3).add_to(mh)

# add the heatmap. The core parameters are:
#--data: a list of points of the form (latitude, longitude) indicating locations of ...
#--radius: how big each circle will be around each Starbucks store
#--blur: the degree to which the circles blend together in the heatmap

mh.add_children(plugins.HeatMap(data=dfm2[['lat', 'lng']].as_matrix(), radius=25, blur=10))
```

```python
mh.save('mapa_calor_saldo_no_castigado.html')
```
