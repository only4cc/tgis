```python
import pandas as pd
import json
import folium
from folium import plugins
```

```python
fn = 'data/' + 'chile.json'
with open(fn, 'r') as datafile:
    data = json.load(datafile)

df = pd.DataFrame(data) 
```

```python
df.head(3)
```

```python
p = [df['lat'][:1], df['lng'][:1]] 

m = folium.Map(
    location=p,
    zoom_start=9
)

tooltip = 'click para info!'
for i in range(1,20): #len(df)):
    x = float(df.iloc[i]['lat'])
    y = float(df.iloc[i]['lng'])
    p = [x, y]     
    folium.Marker(p, popup='<i>'+df.iloc[i]['name']+'</i>', tooltip=tooltip).add_to(m)
    folium.CircleMarker(p, radius=3, weight=2, color='red', fill_color='red', fill_opacity=.5).add_to(m)
```

```python
for i,row in df.iterrows():
    folium.CircleMarker((row.lat,row.lng), radius=3, weight=2, color='red', fill_color='red', fill_opacity=.5).add_to(m)

# add the heatmap. The core parameters are:
#--data: a list of points of the form (latitude, longitude) indicating locations of ...
#--radius: how big each circle will be around each Starbucks store
#--blur: the degree to which the circles blend together in the heatmap

m.add_children(plugins.HeatMap(data=df[['lat', 'lng']].as_matrix(), radius=25, blur=10))
```

```python
m.save('mapa_calor_1.html')
```
